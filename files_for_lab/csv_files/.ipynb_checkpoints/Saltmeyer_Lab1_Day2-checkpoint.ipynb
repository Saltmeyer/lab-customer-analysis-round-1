```python
file1_df = file1_df[['Customer', 'State', 'Customer Lifetime Value', 'Education', 'Gender','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Total Claim Amount', 'Vehicle Class']] 
```

```python
file2_df = file2_df[['Customer', 'State', 'Customer Lifetime Value', 'Education', 'Gender','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Total Claim Amount', 'Vehicle Class']]
```

```python
file_m = pd.concat([file1_df, file2_df, file3_df], axis=0)
```

```python
file_m.columns
```

```python
display(file_m)
```

```python
# Which columns are numerical / categorical?
file_m.dtypes
# Income, Monthly Premium Auto and Total Claim Amount are numerical as they are floats. The rest are objects, i.e. strings
```

```python
# Delete the column education and the number of open complaints from the dataframe.
file_m = file_m.drop('Education', 1) # Executed this again when already deleted. I believe this is why the error shows.
```

```python
file_m = file_m.drop('Number of Open Complaints', axis=1)
```

```python
display(file_m)
```

```python
file_m['Customer Lifetime Value'] = file_m['Customer Lifetime Value'].str.replace('%',' ')
# Source: https://datatofish.com/replace-character-pandas-dataframe/
```

```python
display(file_m)
```

```python
# Changing Customer Lifetime Value to numeric
```

```python
file_m ['Customer Lifetime Value'] = pd.to_numeric(file_m['Customer Lifetime Value'], errors='coerce')
# data['ic5'] =  pd.to_numeric(data['ic5'], errors='coerce')
```

```python
file_m.dtypes
```

```python
# multiply by 100: df['a'] = df['a'].apply(lambda x: x + 1) from: https://stackoverflow.com/questions/34962104/how-can-i-use-the-apply-function-for-a-single-column
file_m['Customer Lifetime Value'] = file_m['Customer Lifetime Value'].apply(lambda x: x*100)
```

```python
display(file_m)
```

```python
# Removing duplicates 
file_m = file_m.drop_duplicates()
```
